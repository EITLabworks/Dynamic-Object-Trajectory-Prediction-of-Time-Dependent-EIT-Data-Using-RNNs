**Loading pretrained models**

```python
from src.lstm_mapper3D_model import mapper3D_model
lstm_mapper = mapper3D_model()
lstm_mapper.summary()
lstm_mapper.compile(optimizer="adam", loss="mse")
lstm_mapper.load_weights("weights/3D_lstm_11200_Trichter_Helix.weights.h5")

from src.no_lstm_mapper3D_model import no_lstm_mapper3D_model
no_lstm_mapper = no_lstm_mapper3D_model()
no_lstm_mapper.summary()
no_lstm_mapper.load_weights("weights/3D_no_lstm_11200_Trichter_Helix_1.weights.h5")

print("Gewichte erfolgreich geladen.")
```

```python
beta =  1.04
vae = vae_model(input_shape=(32, 32, 32, 1), filters=[1, 4, 8, 16], beta=beta)
vae.build(input_shape=(32, 32, 32, 1))
vae.load_weights("weights/3Dvae_best.weights.h5")
print("Gewichte erfolgreich geladen.")
```

**Loading test sim data**

```python
voltage_t, gamma_true, pos, temp, time  = load_3Ddata("exp_data_set_3D_1","exp_data_3D6", use_mean= True)

#Anomalie 0.2 Testdata:
#####################################################################
# exp_data_set_3D/exp_data_3D18 - Helix, 0.7, 11200, n = 20
# exp_data_set_3D/exp_data_3D16 - Helix, 0.5, 8000, n = 20
# exp_data_set_3D/exp_data_3D23 - Helix, 0.6, 9600, n = 20
# exp_data_set_3D_1/exp_data_3D1 - Helix, 0.7, 1400, n = 20 (achtfache Geschwindigkeit)
# exp_data_set_3D_1/exp_data_3D6 - Circularsinwave, 0.7, 3000 
#####################################################################
```

**Reconstruction using voltage data**

```python
plot_temperature_vs_time(time, temp, fpath = "Abbildungen_Temp/", fname = "schnelle_Helixbahn", save = False)
```

```python
model_variant='3D_lstm_small_helix'
lstm = False

if lstm == True:
    pred_latent = lstm_mapper.predict(voltage_t)
else: 
    pred_latent = no_lstm_mapper.predict(voltage_t)

gamma_true_pred = vae.decoder.predict(pred_latent)
```

```python
gamma_true_pred = np.squeeze(gamma_true_pred, axis=-1)

gamma_true_pred = np.clip(gamma_true_pred, a_min =0, a_max=1)
```

```python
from src.performance_eval_2D import plot_voxel_overlay, plot_position_error, plot_volume_deviations

plot_voxel_overlay(gamma_true, gamma_true_pred, pos, save = False, fpath='Abbildung_3D/', fname=f'seq_recon_overlay_{model_variant}')
```

```python
plot_position_error(gamma_true_pred, gamma_true,  fpath='Abbildung_3D/',
    fname=f'seq_recon_{model_variant}', save = False)
```

```python
plot_volume_deviations(gamma_true_pred, gamma_true, fpath='Abbildung_3D/', fname=f'seq_recon_{model_variant}', save = True)
```
