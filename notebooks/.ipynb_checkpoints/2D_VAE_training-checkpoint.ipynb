```python
import numpy as np
import matplotlib.pyplot as plt
from pyeit import mesh
import os
from glob import glob

os.environ["CUDA_VISIBLE_DEVICES"] = "0"

import tensorflow as tf
from src.vae_model import vae_model
from sklearn.model_selection import train_test_split
from src.util import plot_mesh_permarray, plot_mesh, load_sim_data
from pyeit.mesh.wrapper import PyEITAnomaly_Circle
import os
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import Mean
from keras import backend as K
import keras_tuner as kt
from tqdm import tqdm
import pandas as pd
import seaborn as sns
```

```python
n_el = 32
mesh_obj = mesh.create(n_el, h0=0.05)

def createAnomaly(ms, x, y, r):
    """Simple function for adding anomaly regions"""
    anomaly = [
        PyEITAnomaly_Circle(center=[x, y], r=r, perm=0.9),
    ]
    ms = mesh.set_perm(mesh_obj, anomaly=anomaly, background=0.1)

    return ms

ms = createAnomaly(mesh_obj, 0.5, 0.5, 0.2)
#plot_mesh(ms)

size = 6000
obj_rs = np.random.choice([0.1, 0.15, 0.2], size)

r_pos = np.random.uniform(0, 0.8, size)
a_pos = np.random.uniform(0, 2 * np.pi, size)

x_pos = r_pos * np.cos(a_pos)
y_pos = r_pos * np.sin(a_pos)

perms = list()
for x, y, r in zip(x_pos, y_pos, obj_rs):
    ms = createAnomaly(mesh_obj, x, y, r)
    ms.perm = ms.perm + np.random.normal(scale=0.05, size=ms.perm_array.shape[0])
    perms.append(ms.perm_array)
    # plot_mesh(ms)
gamma = np.array(perms)
```

```python
from src.util import create_trajectory

def plot_mesh_with_training_pos(mesh_obj, positions_x, positions_y,fpath='', fname = '',save=False):

    center_pos = create_trajectory("Quadrat", 0.5, 1000)
    
    fig, ax = plt.subplots(figsize=(8, 8))
    
    pts = mesh_obj.node
    tri = mesh_obj.element
    
    ax.triplot(pts[:, 0], pts[:, 1], tri, 'k-', alpha=0.2, lw=0.5)
    
    el_x = pts[mesh_obj.el_pos, 0]
    el_y = pts[mesh_obj.el_pos, 1]
    ax.plot(el_x, el_y, "ro")

    #for i in range(len(mesh_obj.el_pos)):
       # ax.text(el_x[i], el_y[i], str(i + 1), size=12)
    
    ax.scatter(positions_x, positions_y, s=50, facecolors='none', 
               edgecolors='blue', label="Positionen")
    
    ax.plot(center_pos[:, 0], center_pos[:, 1], 'r--', linewidth=1, label="Trajektorie")
    ax.scatter(center_pos[:, 0], center_pos[:, 1], s=10, color='red', alpha=0.5)

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_aspect("equal")
    ax.set_ylim([-1.2, 1.2])
    ax.set_xlim([-1.2, 1.2])
    
    plt.tight_layout()

    if save:
        os.makedirs(fpath, exist_ok = True)
        full_path = os.path.join(fpath, fname)

    if not full_path.endswith('.pdf'):
        full_path += '.pdf'

        fig.savefig(full_path, format = 'pdf', bbox_inches = 'tight', dpi =300)
    
    return fig, ax

fig, ax = plot_mesh_with_training_pos(mesh_obj, x_pos, y_pos, fpath= 'NeueAbbildung/', fname='points_square', save = True)
plt.show()
```

```python
def plot_random_samples(mesh_obj, gamma_array, num_samples=20):
    
    
    random_indices = np.random.choice(len(gamma_array), size=num_samples, replace=False)
    
    # Subplots erstellen
    cols = 5
    rows = int(np.ceil(num_samples / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 3))
    axes = axes.flatten()
    
    for i, idx in enumerate(random_indices):
        ax = axes[i]
        plot_mesh_permarray(mesh_obj, gamma_array[idx, :], ax=ax, title=f"Sample {idx}")
    
    # Überflüssige Subplots entfernen
    for j in range(num_samples, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

plot_random_samples(mesh_obj, gamma, num_samples=20)
```

def save_statistics(data_dict, filepath='statistics.txt'):
    """
    Speichert die Statistiken der Abweichungen in einer Textdatei.
    
    Args:
        data_dict: Dictionary mit den x-Abweichungen, y-Abweichungen und Perm-Werten
        filepath: Pfad und Name der zu erstellenden Textdatei
    """
    # Bereinige die Daten von NaN-Werten
    cleaned_dict = clean_data_dict(data_dict)
    
    # Berechne die Statistiken
    x_mean = np.mean(cleaned_dict['x-Abweichung'])
    y_mean = np.mean(cleaned_dict['y-Abweichung'])
    perm_mean = round(np.mean(cleaned_dict['Perm']))
    st_fe = 2807
    perm_percentage = round((perm_mean/st_fe)*100, 2)
    
    # Erstelle den Text mit den Statistiken
    statistics_text = (
        f"Auswertung der Abweichungen\n"
        f"===========================\n\n"
        f"Durchschnittliche x-Abweichung: {x_mean:.2f}\n"
        f"Durchschnittliche y-Abweichung: {y_mean:.2f}\n"
        f"Mittlere Perm-Abweichung: {perm_mean} [FE]\n"
        f"Prozentuale Abweichung: {perm_percentage} [%]\n"
    )
    
    # Speichere die Statistiken in einer Textdatei
    with open(filepath, 'w', encoding='utf-8') as f:
        f.write(statistics_text)
    
    print(f"Statistiken wurden in '{filepath}' gespeichert.")
